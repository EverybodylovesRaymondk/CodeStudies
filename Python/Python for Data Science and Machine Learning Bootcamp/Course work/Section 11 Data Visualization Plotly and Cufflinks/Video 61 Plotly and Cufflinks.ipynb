# Creating Data

```python
df=pd.DataFrame(np.random.randn(101,4),columns='A B C D'.split())
```

```python
df.head()
```

```python
df2 = pd.DataFrame({'Category':['A','B','C'],'Values':[32,43,50]})
```

```python
df2
```

# Creating a line plot 

```python
df.iplot(xTitle='Index',yTitle='Value',title='Line plot Df')
```

# Creating a scatter plot

```python
#Note: Mode='markers' is needed because by default iplot will connect each value in the data set with a line
df.iplot(kind='scatter',x='A',y='B',mode='markers',size=5,xTitle='A',yTitle='B')
```

# Bar Plot

```python
df2.iplot(kind='bar',x='Category',y='Values',title='Bar plot of DF2 ')
```

```python
df.sum().iplot(kind='bar',title='Bar of sum for each column of df')
```

# Box Plot

```python
df.iplot(kind='box')
```

# 3D surface plot

```python
df3=pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[5,4,3,2,1]})
```

```python
df3
```

```python
df3.iplot(kind='surface',colorscale='rdylbu')
```

# Histogram

```python
#Creating histogram plot for column A of df
df['A'].iplot(kind='hist',bins=50)
```

# Spread

This will be usefull for stock / forex plotting. 

The spread is the differnce between the values plotted.

```python
df[['A','B']].iplot(kind='spread')
```

# Bubble plot

```python
df.iplot(kind='bubble',x='A',y='B',size='C')
```

# Scatter Matrix

Essentially the same thing as a pair plot

```python
df.scatter_matrix()
```
