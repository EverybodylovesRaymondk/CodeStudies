```python
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

# Setting the background of the plot

### Style

```python
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
```

### Removing the plot spines

```python
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
#Removing spines , note by default top and right is true but left and bottom can be added as arguments
sns.despine(left=True,bottom=True)
```

### Setting figure size for seaborn plots using matplot plt

```python
plt.figure(figsize=(12,5))
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
sns.despine(left=True,bottom=True)
```

### Setting the context and font scale of a plot

```python
sns.set_context('notebook',font_scale=2)
sns.countplot(x='sex',data=tips)
```

### Pallets and colors

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='magma')
```
