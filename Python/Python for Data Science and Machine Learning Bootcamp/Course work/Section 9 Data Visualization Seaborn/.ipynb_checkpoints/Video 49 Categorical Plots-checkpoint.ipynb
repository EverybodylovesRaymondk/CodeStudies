```python
import seaborn as sns
import numpy as np
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

# Bar Plot

The bar plot takes in a categorical and a numerical value and by default will plot the mean of the numerical value in relation to the categorical value

```python
#Creating a Bar plot of the mean average total bill for each type of sex in the dataset
sns.barplot(x='sex',y='total_bill',data=tips)
```

## Bar Plot estimator function

By default as stated above the barplot will by default show the mean average. This can be changed using the estimator argument to example the standard deviation or any function (custom functions)

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator = np.std)
```

# Count Plot

The count plot only counts the number of instances of the categorical value in the data set. Requiring only the x argument (categorical value)

```python
sns.countplot(x='sex',data=tips)
```

# Box Plot

The box plot will show the inter quartile ranges for the total bill spread over the different days in the dataset

```python
#Creating a box plot (x=categorical value) , (y=numerical value) that shows the inter quantile range plus out liers
sns.boxplot(x='day',y='total_bill',data=tips)
```

Adding the hue argument allows us to split the plot with another categorical value (smoker) for example

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

# Violin plots

Shows the distrubution of the data over a category. The vilolin plot shows the distrubution of the points on its side essentially the kde of the underlying distribution of each category 

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

Adding the hue and split aguments to the violin plot allow the data to be shown for 2 categories on the same plot 

The plot below shows the distribution of the total bill for a day split into male and female

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

# Strip Plot

Stip plot will create a scatter plot of the numerical values based on the categorical argument 

```python
#Adding jitter will spread out the plot values to better represent the number of data points on a given level
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex')
```

# Swarm plot

Simular to a strip plot except that the points are adjusted that they do not over lap

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex')
```

## Combining a swarm plot with a violin plot

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
sns.swarmplot(x='day',y='total_bill',data=tips,color='lime')
```

# Factor plot

### This is a generic plot type where the kind argument is used to dictate the kind of plot to create. 

#### Note: Factor plot has been renamed to catplot in the new version of seaborn

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='boxen')
```
