**K Means Clustering**
This model tries and cluster together data that is unlabled. 
It does this by looking for common features of data and tries to determin a common group for the data point

# Importing Standard Modules

```python
import pandas as pd
import numpy as np
from termcolor import colored
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
#Make_blobs allows you to create sample data using seaborn that can be used to test clustering models
from sklearn.datasets import make_blobs
```

# Creating random data

```python
data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
type(data[0])
```

Note: data is a tuple where each value is a numpy array

# Plotting the data

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

# Importing KMeans clustering model

```python
from sklearn.cluster import KMeans
```

# Creating and fitting a k means model

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
kmeans.cluster_centers_
```

# Plotting "Predicted" clustors vs actual

```python
fig , (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title('K_Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

The color of the groups does not matter. 

The observation can be made on individual points where the model did not correctly "Predict" the clustor for that data point.
