# Build in Pandas plots

### Histogram

```python
sns.set_style('white')
sns.despine(left=True,bottom=True)
df1['A'].hist(bins = 30)
```

### Area Plot

```python
df2
```

```python
#Starting from index 0 the graph is plotted for each column in the data frame. 
#Alpha argument is used to effect the opticality of the plot. 
df2.plot.area(alpha=0.2)
```

### Bar Plot

```python
#Each bar is representative of the value for each column at that index e.g index 0 has values for a-d
#Using the stacked argument places eacy value on top of the other in the series
df2.plot.bar(stacked=True)
```

### Line Plot

```python
#plotting the values in column b for the index (dates)
#Note the x axis defaults to the index
#Addding figsize argument allow you to increase width and height of the plot
#Adding lw allows you to set the line width. 
df1.plot.line(y='B',figsize=(20,6),lw=0.5)
```

### Scatter Plots

```python
#Plogging the values of A and B making the plot size of the items based of column 'C' * 100 and the color is of column D
df1.plot.scatter(x='A',y='B',s=df1['C']*50,c='D')
```

### Box Plot

```python
#Creating box plot of the values for every column in dataframe 2 
df2.plot.box()
```

### Bi Variable data

```python
dfn = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
```

### Hexbin plot

```python
dfn.plot.hexbin(x='a',y='b',gridsize=25,cmap='magma')
```

### Kernal density estimation (KDE) Plot

```python
#Creating a kde plot for every column in df2
df2.plot.kde()
```
