```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y = x **2
```

```python
y
```

# Crating a plot using the plot function

```python
#Creating the plot
plt.plot(x,y)
#Setting the axis lables
plt.xlabel('X Vale')
plt.ylabel('Y Value')
#Setting plot title
plt.title('Plot title')
```

Creating subplots multiple plots on the same plot out line

```python
plt.subplot(1,2,1) #The values represent (number of rows , number of columns , which position) the plot must be in 
plt.plot(x,y,'r') #will plot red lind

plt.subplot(1,2,2)
plt.plot(y,x,'b') # will plot blue line
```

# Using object oriented method to plot

```python
fig = plt.figure()
#Adding axis to the figure object (Canvas) Left , Bottom , With ,  Heigt that ranges from 0-1
fig = fig.add_axes([0.1,0.1,0.8,0.8])
#Plotting the values onto the axes
fig.plot(x,y)
#Setting axis lables and title
fig.set_xlabel('X Lable')
fig.set_ylabel('Y Lable')
fig.set_title('Title')
```

Creating a plotting figure (canvas) with multiple axises

```python
fig = plt.figure()
#Adding axis to the figure object (Canvas) Left , Bottom , With ,  Heigt that ranges from 0-1
figAx1 = fig.add_axes([0.1,0.1,0.8,0.8])
figAx2 = fig.add_axes([0.2,0.5,0.4,0.3])

figAx1.plot(x,y,'r')
figAx1.set_title('Large Plot')

figAx2.plot(y,x,'b')
figAx2.set_title('Small Plot')
```
