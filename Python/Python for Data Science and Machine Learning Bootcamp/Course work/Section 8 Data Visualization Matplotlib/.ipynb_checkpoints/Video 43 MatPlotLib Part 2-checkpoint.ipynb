```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x=np.linspace(0,5,11)
y = x **2
```

# Creating sub plots using Object orientated figures

```python
fig,axes=plt.subplots(nrows=1,ncols=3)
#Adjusting layout so plots does not over lap
plt.tight_layout()

#Note that axes is a array that can be itterated through or indexed

#Plotting with itteration

# for current_ax in axes:
#     current_ax.plot(x,y)

#Plotting with indexing
axes[0].plot(x,y,'r')
axes[0].set_title('First Plot')

axes[1].plot(y,x,'b')
axes[1].set_title('Second Plot')

plt.show()
```

# Figure Size and DPI

```python
#Creating plot figure
#Figuresize is a tuple (with,height) in inches
fig = plt.figure(figsize=(8,2),dpi=150)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
#Creating subplot
#Figuresize is a tuple (with,height) in inches
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2),dpi=100)
plt.tight_layout()
axes[0].plot(x,y)
axes[1].plot(y,x,'r')
```

# Saving a figure (plot) as a file

```python
fig.savefig('my_picture.png',dpi=200)
```

# Adding legends to plots

```python
figure = plt.figure(figsize=(4,4),dpi=200)
ax = figure.add_axes([0.1,0.1,0.8,0.8])
#The label argument in the plot call enables the legend on the plot
ax.plot(x,x**2,label = 'X Squared')
ax.plot(x,x**3,label = 'X Cubed')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_title('Plot with legend')
ax.legend()
plt.show()
```

```python
figure.savefig('Plot with legend.png',dpi=200,)
```
