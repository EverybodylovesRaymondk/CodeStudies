```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x=np.linspace(0,5,11)
y = x **2
```

# Plot Appearance

Changing colors of the lines in a plot
#NB LOOK AT THE MATPLOTLIB COURSE NOTES FOR ADDITIONAL PLOTS

```python
fig1 = plt.figure()

ax = fig1.add_axes([0,0,1,1])

#common colors can be used by name
ax.plot(x**2,y**0.9 , color = 'green')

#a color can also be set by RGB Hex Codes
ax.plot(y,x , color = '#FF8C00')
```

# Changing line width (lw) and tranparancy of a line(alpha)

```python
fig2 = plt.figure()

ax = fig2.add_axes([0,0,1,1])

ax.plot(x,y , color = 'purple' , lw = 5 , alpha = 0.3)
```

# Changing the line style(ls) of the line

```python
fig3 = plt.figure()
ax = fig3.add_axes([0,0,1,1])

ax.plot(x*y,y , color = 'Lime' , lw = 3 , ls='--')
ax.plot(x*x,y , color = 'red' , lw = 3 ,  ls='-.')
```

# Adding markers on a line

```python
fig4 = plt.figure()
ax = fig4.add_axes([0,0,1,1])

ax.plot(x,y , color = 'red' , lw = 1, marker = '*')
ax.plot(x*y,y , color = 'blue' , lw = 1, marker= 'o')
```

# Customizing line markers

```python
fig5 = plt.figure()
ax = fig5.add_axes([0,0,1,1])

ax.plot(x,y , color = 'red' , lw = 3, marker = 'o',
       markersize = 20,
       markerfacecolor = 'blue',
       markeredgewidth = 3,
       markeredgecolor = 'lime')
```

# Setting X / Y axis limits

```python
fig5 = plt.figure()
ax = fig5.add_axes([0,0,1,1])
ax.plot(x,y , color = 'red' , lw = 1, marker = '*')
ax.set_xlim([0,1])
ax.set_ylim([0,7])
```
