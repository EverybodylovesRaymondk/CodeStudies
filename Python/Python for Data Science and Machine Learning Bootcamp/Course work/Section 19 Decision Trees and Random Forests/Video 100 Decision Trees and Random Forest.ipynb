# Importing standar modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Importing ML modules

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
# Randomforest
from sklearn.ensemble import RandomForestClassifier
```

# Importing evaluation modules

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
df = pd.read_csv(DLoc)
```

```python
#Note : age is in months
df.head()
```

## Visualizing correlations in the data with pairplot

```python
sns.pairplot(df,hue='Kyphosis')
```

# Splitting data for training

```python
X = df.drop('Kyphosis',axis=1)
y = df['Kyphosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=101)
```

# Training a single decision tree

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

# Creating predictions

```python
predictions = dtree.predict(X_test)
```

# Evaluating the ML model's predictions

```python
print(confusion_matrix(y_test,predictions))
print('\n')
print(classification_report(y_test,predictions))
```

# Creating a Random forest ML model

```python
rfc = RandomForestClassifier(n_estimators=24)
```

```python
rfc.fit(X_train,y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

# Evaluating the Random Forces Clasifier model

```python
print(confusion_matrix(y_test,rfc_pred))
print('\n')
print(classification_report(y_test,rfc_pred))
```

# Using the elbo methot to hunt for the optimal number of estimators

```python
#Creating empty list to store error rates
error_rate = []
#Creating a for loop to use a range 1 to 40 as the n_neighbors values and create predictions for each value of i
for i in range(1,200):
    RFC = RandomForestClassifier(n_estimators=i)
    RFC.fit(X_train,y_train)
    pred_i = RFC.predict(X_test)
#appending the mean value where the predicted value was not equal to the actual values(mean of the error)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,200),error_rate,color='blue',linestyle='dashed',marker='o',
        markerfacecolor='red',markersize=10)
plt.title('Error Rate VS N Value')
plt.xlabel('N')
plt.ylabel('Error Rate')
```
